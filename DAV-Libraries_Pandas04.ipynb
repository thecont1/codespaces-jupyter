```python
import pandas as pd
import numpy as np

movies = pd.read_csv('movies.csv', index_col=0)
directors = pd.read_csv('directors.csv', index_col=0)

data = movies.merge(directors, how='left', left_on='director_id', right_on='id')
data.drop(['director_id','id_y'], axis=1, inplace=True)
```

```python
data
```

```python
data.groupby(['director_name'])['title'].count().sort_values(ascending = False)
```

```python
data_agg  = data.groupby(['director_name'])[["year", "title"]].agg( {"year": ['min', 'max'], "title":"count"})
```

```python
data_agg.columns
```

```python
data_agg
```

```python
data_agg["year"]
```

```python
data_agg = data.groupby(['director_name']).aggregate(
    {"year":['min', 'max'], "title": "count"})

data_agg
```

```python
data_agg.columns
```

```python
new_column_names = ['year_min', 'year_max', 'title_count']
```

```python
# new_column_names = ['_'.join(col) for col in data_agg.columns]

# new_column_names
```

```python
data_agg.columns = new_column_names
```

```python
data_agg
```

```python
data_agg.reset_index()
```

```python
data_agg['yrs_active'] = data_agg['year_max'] - data_agg['year_min']
```

```python
data_agg
```

```python
data_agg["movie_per_yr"] = data_agg["title_count"]/data_agg["yrs_active"]
```

```python
data_agg
```

```python
data_agg.sort_values(by = "movie_per_yr", ascending = False)
```
